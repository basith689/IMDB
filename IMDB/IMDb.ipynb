# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
movies = pd.read_csv('imdb_movies_.csv')
movies.head()
```

```python
# 1. Descriptive Statistics
total_movies = len(movies)
year_range = (movies['Year'].min(), movies['Year'].max())
average_rating = movies['Rating'].mean()
top_5_movies = movies.nlargest(5, 'Rating')[['Title', 'Rating']]

print(f"Total Movies: {total_movies}")
print(f"Year Range: {year_range}")
print(f"Average Rating: {average_rating}")
print(f"Top 5 Movies:\n{top_5_movies}")
```

```python
# 2. Rating Distribution
plt.figure(figsize=(10, 6))
sns.histplot(movies['Rating'], bins=20, kde=True)
plt.title("IMDb Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Content Rating Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=movies, x='Content Rating', order=movies['Content Rating'].value_counts().index)
plt.title("Content Rating Distribution")
plt.xlabel("Content Rating")
plt.ylabel("Count")
plt.show()
```

```python
# 4. Yearly Trend
movies_per_year = movies.groupby('Year').size()
plt.figure(figsize=(12, 8))
movies_per_year.plot(kind='line')
plt.title("Number of Movies Released Each Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Box Plot for Ratings by Content Rating
plt.figure(figsize=(12, 8))
sns.boxplot(data=movies, x='Content Rating', y='Rating')
plt.title("IMDb Ratings by Content Rating")
plt.xlabel("Content Rating")
plt.ylabel("Rating")
plt.show()
```

```python
# 6. Scatter Plot: Year vs Rating
plt.figure(figsize=(12, 8))
sns.scatterplot(data=movies, x='Year', y='Rating')
plt.title("Year vs IMDb Rating")
plt.xlabel("Year")
plt.ylabel("Rating")
plt.show()
```
